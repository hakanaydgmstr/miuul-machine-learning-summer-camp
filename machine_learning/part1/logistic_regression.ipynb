# Logistic Regression

Amaç sınıflandırma problemi için bağımlı ve bağımsız değişkenler arasındaki ilişkiyi doğrusal olarak modellemektir.

Yukarıdaki z formülü, bizim lineer regresyondan alışık olduğumuz "*y = b + w.x*" formülüdür. Burada farklı olarak bu formülden bulacağımız sonucu, görseldeki üst formülde (sigmoid fonksiyonu) yerine yazarak ekstra bir işlem ile, bir sınıfa ait olma olasılıklarını inceliyor olacağız.


Lojistik regresyon bir sınıflandırma modelidir. Bağımlı değişkenin 0 veya 1 değerini aldığı senaryolarda, modelin tahminin 0 veya 1 olmasını bekleriz. Fakat lineer regresyon denklemini hesapladığımızda elde edeceğimiz sonuç 0.45 olabilir, -0.8 olabilir, 1.35 veya -1.4 olabilir. Bu hesaplanan değerleri de 0 ve 1 arasında bir değere dönüştürmek için *sigmoid fonksiyonunu* kullanıyoruz. Bu fonksiyonun ürettiği değer bir olasılık değeridir.

Yukarıdaki grafik sigmoid fonksiyonu grafiğidir. Hesapladığımız z değişkeni x-ekseni üzerinde bir değer alacaktır. Buradaki değer ile grafikteki eğri yardımıyla y-eksenindeki değer okunur ve böylece olasılık hesaplanmış olur. Örnek olarak sol tarafta yazan denklem, X değişkeninin x olduğu durumda Y değişkeninin 1 olma olasılığını ifade eder. Örneğin rastgele bir hesap sonucu 0.74 sonucunu elde ederiz. Bu, 1 sınıfına ait olma olasılığıdır (soldaki denkleme göre). Daha sonra bu olasılıkları biz bir eşik değere göre dönüştürürüz.

Verilen değerlere sahip bir gözlem biriminin 1 sınıfına ait olma olasılığı 0.731miş. Bizim bu modelden tahmin çıktısı olarak 1 veya 0 elde etmemiz gerekiyordu. Dolayısıyla belirlediğimiz bir eşik değerine göre elde ettiğimiz sonucu bir sınıfa sokabiliriz. Örnek olarak eşik değerini 0.5 olarak belirleyelim. Hesapladığımız değer 0.5'in yukarısında olursa o gözlem birimi 1 sınıfına, aşağısında olursa 0 sınıfına aittir yorumunu yapabiliriz.

Peki bu işlemler nasıl gerçekleştirilir? Bir optimizasyon problemi olarak konu ele alındığında bu problem nasıl çözülür?

Gerçek değer ile tahmin edilen değerler arasındaki farklara ilişkin *Log Loss* fonksiyonunu minimum yapabilecek ağırlıkları bularak.

Linear Regression'da MSE adında bir fonksiyonumuz vardı. Bu fonksiyon minimum olacak şekilde ağırlıkları buluyorduk. Buradaki işlem de benzer şekilde işlemektedir. *Log Loss* fonksiyonu minimum olacak şekilde ağırlıkları bulmaya çalışıyoruz.

### Confusion Matrix (Karmaşıklık Matrisi)

Confusion matrix'de tahmin edilen değerler bir matris içerisine, doğru veya yanlış tahminlerine göre kaydedilirler.

Yukarıdaki örnekte, 1 ve 0 sınıflarından oluşan bir bağımlı değişkenimiz olduğunu düşünelim. Oluşturacak olduğumuz matrisin satırları bu gerçek sınıfları, sütunları ise tahmin edilen sınıfların değerini tutacak. Her bir sınıfın birbirleri ile kesiştiği noktalarda birer ifade yer alıyor. Örnek olarak true pozitif değeri, 1 sınıfının içerisinde yer alan bir gözlem birimini 1 olarak tahmin ettiğimiz için true değerini aldı, ve tahmin ettiğimiz sınıf 1 olduğu için de pozitif olmuş oldu.

Bir diğer örnek false negatif'te, 1 sınıfına ait olan bir gözlem birimini 0 olarak tahmin ettiğimiz için false ibaresini alır, ve yine aynı şekilde tahmin ettiğimiz değer 0 olduğu için negatif olmuş oldu. Bu mantık ile true negatif bölümünü değerlendirdiğimizde, 0 sınıfına ait gözlem birimlerini 0 olarak tahmin edilen değerlerin sayısını tutacaktır.

Özet olarak, true olan yerlerde doğru, false gördüğüm yerlerde yanlış tahminlerde bulunmuşuz oluruz. Negatifler 0 sınıfının tahmin edildiğini, pozitifler ise 1 sınıfının tahmin edildiğini temsil eder.

**-Accuracy:** Doğru sınıflandırma oranıdır. Doğru yapılan işlemlerin toplamlarının, toplam tahmin sayısına bölümüdür.

(TP+TN) / (TP + TN + FP + FN)

**-Precision:** Pozitif sınıf (1) tahminlerinin başarı oranıdır. Sınıf tahminlerine odaklanılır! (1 sınıfını tahmin ettiklerimde ne kadar başarılıyım?) (1 sınıfını tahmin ettiklerimin gerçekte ne kadarı 1 sınıfına ait?)

TP / (TP + FP)

**-Recall:** Pozitif sınıfın (1) doğru tahmin edilme oranıdır. Tahminlere değil gerçek değerlere odaklanılır! (1 sınıfı olan gözlem birimlerini doğru tahmin etme oranı)(1 sınıfına ait olan gözlemlerin ne kadarının doğru tahmin edildiğini ifade eder.)

TP / (TP + FN)

 False Pozitif (FP) üretmek yanlış alarm üretmek demektir (1. Tip hata).
 
 False Negatif (FN) kritik hatadır! (2. Tip hata)

**-F1 Score:** Bu score hem precision hem de recall değerlerinin etkilerini kendi içerisinde tutmaktadır.

2 * (Precision * Recall) / (Precision + Recall)

Eğer elimizdeki sınıflandırma problemi dengeli sınıf dağılımına sahipse **accuracy**'i verebiliriz. Ama eğer elimizdeki veri setinde sınıflar dengesiz dağılıyorsa bu durumda direkt accuracy oranını kullanamayız (bizi yanıltabilir). **Recall**, **precision** ve **F1 score** değerlerine bakmamız gerekir. 

### Classification Threshold

Herhangi bir sınıflandırma probleminde bir model kurduğumuzu düşünelim. Bu model neticesinde 1. sınıfa ait olma olasılıklarını buluruz. **Dikkat!** Kullandığımız fonksiyonlardan dolayı bu olasılıkları görmüyor olsak bile arka planda her zaman 1 sınıfına ait olma olasılığı bulunur.

Fakat bu tür sınıflandırma modellerinde elde ettiğimiz tahminler her zaman 1 ve 0'lardan oluşmaktaydı. Modellerden elde ettiğimiz olasılık değerleri 1 ve 0 sınıflarına nasıl dönüştürüldü? Bunun sebebi *eşik değerlerdir*. Çalışmanın başında belirtilen eşik değere göre eğer 1 sınıfına ait olma olasılığı belirlenen eşik değerden büyükse, bu sınıfı 1 olarak ata, bu değerden küçükse 0 olarak ata denir. 

Yukarıdaki örnek veri setinde sol sütunda, o gözlem biriminin gerçek değeri, sağ sütunda ise model sonucu o gözlem biriminin 1 sınıfına ait olma olasılıkları verilmiştir. Ortadaki sütun ise, sağ sütundaki olasılıkları belirli bir threshold değerine göre atanmış değerleri barındırıyor. Buradan hareketle threshold değerinin 0.5 olduğu bir senaryoda ulaştığımız accuracy değeri 0.70'tir (Doğru tahmin edilenler / toplam tahminler).

Aynı veri setinde threshold değerini 0.75 olarak belirlediğimizde, tahmin edilen olasılıklar ve sınıf değerleri değişecektir. Yukarıdaki gibi sınıf atama işlemlerinden sonra accuracy hesapladığımızda 0.50 değerini elde etmiş oluruz.

Buradan hareketle, genellikle threshold değerini arttırdığımızda accuracy değerinin de azaldığını gözlemleyebiliriz.

Yine aynı veri seti üzerinden bu sefer threshold'u azalttığımız zaman accuracy değerimizin 0.80lere kadar çıktığını görüyoruz. 

Eşik değerinin değişmesine göre sınıflandırma nihai sınıf tahminlerimiz değişiyor. Demek ki, öyle bir şey olmalı ki, olası bütün eşik değeri değişimleri de göz önünde bulundurularak bir hesaplama yapılsın ve buna göre matematiksel ifade karşımıza gelsin. 

Çalışmalarda threshold değeri default olarak 0.5 belirlenir fakat bu değeri değiştirdiğimizde gözlemliyoruz ki bütün değerler, accuracy, precision, recall hepsi değişmekte. Bu problemi çözmek için *ROC Curve* yöntemi kullanılır. 

### ROC Curve

Yukarıdaki grafiğe roc eğrisi denir. Eğer hiçbir model kurmasaydık ve rastgele 1 ve 0 atamaları yapsaydık bu durumda elde edeceğimiz başarı grafik üzerindeki kesikli kırmızı doğru (random classifier) olacaktı. Yani burası, temel bir karşılaştırma noktasıdır. Bu eksenin yukarısındaki diğer çizgiler yukarı doğru kapladıkları alan arttıkça başarı oranı artıyor olarak düşünülür.

Öncesinde threshold değeri değiştikçe tahmin edilen sınıflar ve istatistik sonuçlarının da değişeceğinden bahsedilmişti. Buradaki amacımız farklı thresholdlar kullanıldığında bunların hangisinin daha başarılı sonuçlar ürettiğini gözlemlemek. Her threshold değiştikçe yeni tahmin elde edilecek, ve yeni tahminler elde edildikçe yeni confusion matrixler oluşturulacak. Bu confusion matrixteki true positive ve false positive rateler yukarıdaki grafikte belirlenerek bir eğri oluşturacaklar. Her bir threshold değerinde bir eğri elde etmiş olacağız. Eğri yukarıya doğru ne kadar alan kaplarsa o threshold o kadar başarılıdır yorumu yapabiliriz.

Bu eğrileri değerlendirebilmek için, eğrinin altında kalan alanın integrali alınırsa bu durumda **area under curve (AUC)** metriği elde edilir. AUC metriği, ROC curve'ün sayısal bir şekilde ifade edilişidir. ROC eğrisinin altında kalan alandır. *AUC*, tüm olası sınıflandırma eşikleri için toplu bir performans ölçüsüdür.

Demek ki sınıflandırma problemleri söz konusu olduğunda dikkat etmemiz gereken birinci konu sınıf dağılımı dengesiz mi? (95'e 5, 80'e 10 vs. Yüzde ile beraber frekans da önemlidir!) Eğer veri dengesiz ise bu durumda recall değerine, precision değerine ve f1 score'una bakıyoruz. F1 score'una veri dengeli de olsa dengesiz de olsa bakıyoruz. Bir de AUC değerlerine bakarız. AUC, olası eşik değerlerine karşılık elde edilecek başarı ölçütüdür. 

### LOG Loss

Optimize etmek için. Log Loss ifadesi bir başarı metriğidir. Bu başarı metriğini hem modelin başarısı için değerlendirebiliriz, hem de optimizasyon yapmak üzere, gerek hiperparametrelerde gerek ağırlıklarda, optimizasyon yapmak üzere ağırlıkları bulurken iterasyon sayısı açısından ya da diğer bazı olası farklı sınıflandırma algoritmalarında hiperparametreleri vs. bulmak adına optimize etmek üzere odaklanacağımız fonksiyondur. 

Bu metrik *cross entropy* metriğidir. 

Yukarıdaki cross entropy, binary classification, yani ikili sınıflandırma problemleri için kullanılan formu.

Yukarıdaki veri seti üzerinden işlem yapacak olursak, sol sütunda gerçek değerler, sağdaki sütunda 1 sınıfına ait olma olasılığı, ortadaki sütunda ise belirli bir eşik değerine göre sağdaki olasılıkların bir sınıfa atanmış halleri bulunmakta.

Formül incelendiği zaman, yi dediğimiz değişken bizim gerçek değerimizdir. p(şapkalı yi) ise 1 sınıfına ait olma olasılığıdır. yi'nin 1 veya 0 olma durumunda denklemdeki -'nin sağ ya da solundaki ifade sıfırlanmış olacaktır. Dolayısıyla yukarıda aslında yapılan işlem, her bir gözlem biriminin gerçek değeri 0 ise o gözlemin 0 olma olasılığının logaritması alınarak -1 ile çarpılması ya da 1 ise o gözlemin 1 olma olasılığının logaritması alınarak -1 ile çarpılmasıdır. Her gözlem birimini hesaplandıktan sonra toplayıp, toplam gözlem sayısına böldüğümüzde log loss değerini hesaplamış oluruz.

Ne kadar yüksek bir tahmin değeri varsa bu durumda logaritmadan çıkacak olan değerle -1 çarpılacağından dolayı, o ifade o kadar küçük olacaktır. Benzer şekilde 1 sınıfına ait olma olasılığı ne kadar düşük ise, logaritmasının -1 ile çarpımı da o kadar büyük çıkacaktır (entropi yüksek olur, entropi yüksekse hata yüksek). Olaslık tahminleri üzerinden entropi hesabı yapılır.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report, plot_roc_curve
from sklearn.model_selection import train_test_split, cross_validate


def outlier_thresholds(dataframe, num_col, q1=0.05, q3=0.95):
    quartile1 = dataframe[num_col].quantile(q1)
    quartile3 = dataframe[num_col].quantile(q3)
    iqr = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * iqr
    low_limit = quartile1 - 1.5 * iqr

    # outliers = [dataframe[(dataframe[num_col] < low) | (dataframe[num_col] > up)]]
    return low_limit, up_limit


def check_outlier(dataframe, num_col):
    low_limit, up_limit = outlier_thresholds(dataframe, num_col)

    if dataframe[(dataframe[num_col] < low_limit) | (dataframe[num_col] > up_limit)].any(axis=None):
        return True
    else:
        return False
    

def replace_with_thresholds(dataframe, num_col):
    low_limit, up_limit = outlier_thresholds(dataframe, num_col)

    dataframe.loc[(dataframe[num_col] < low_limit), num_col] = low_limit
    dataframe.loc[(dataframe[num_col] > up_limit), num_col] = up_limit
    

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.width', 500)
```

### Model Evaluation

Elimizde gerçek veriler ve tahmin verileri var. Bunları kullanarak bir başarı değerlendirmesi yapabiliriz.

Confusion matrix'i kullanacağız. Fakat bunu görsel bir şekilde yapmak istediğimiz için bir fonksiyon yazacağız:

```python
def plot_confusion_matrix(y, y_pred):
    acc = round(accuracy_score(y, y_pred), 2)
    cm = confusion_matrix(y, y_pred)
    sns.heatmap(cm, annot=True, fmt=".0f")
    plt.xlabel("y_pred")
    plt.ylabel("y")
    plt.title("Accuracy score: {0}".format(acc), size=10)
    plt.show()
```

```python
plot_confusion_matrix(y, y_pred)
```

### Model Validation: 10-Fold Cross Validation

Burada öyle bir işlem yapmalıyız ki modelin doğrulama sürecini en doğru şekilde ele alabilelim. Bunu yapmanın yolu da *çapraz doğrulama işlemidir*. Yani az önce yapmış olduğumuz holdout işlemini 10 katlı bir şekilde yapacağız.

```python
y = df["Outcome"]
X = df.drop("Outcome", axis=1)
```

**Dikkat!** Burada bütün veriyi kullanarak bu işlemi gerçekleştireceğiz. Yani train ve test olarak ayırmadan. Bu noktada tercih edilmesi gereken iki şey olabilir: Birincisi; veri seti bol ise veri setini baştan holdouttaki gibi train ve test olarak ikiye ayırıp, train setini 10 katlı validasyon yapıp en son bir de test seti performansı incelenebilir. İkincisi; eğer veri seti bol değilse, az sayıda gözlem varsa, bu durumda bütün veriyi kullanarak bu işlemi yapabiliriz.

```python
log_reg = LogisticRegression().fit(X, y)
```

Modelimizi kurduk. Şimdi cross validation'u uygulamamız gerekiyor.

```python
cv_results = cross_validate(log_reg, # modelimiz
                            X, y, # bağımlı ve bağımsız değişkenlerimiz
                            cv=5, # kaç katlı cross validate olacağı
                            scoring=["accuracy", "precision", "recall", "f1", "roc_auc"]) # metodun hesaplamasını istediğimiz değerler
```

5 kat train 1 kat test olarak işlemlerimizi yaptık. cv_result'un içerisinden test_accuracy değerlerini çağırarak bu 5 kat için de hesaplanan accuracy değerlerine bakabiliriz.

```python
cv_results['test_accuracy']
```

Yukarıdaki accuracy'lerin ortalamasını da alabiliriz.

```python
cv_results['test_accuracy'].mean()
```

Bu ortalama sonucu, bizim 10-katlı çapraz doğrulama ile elde etmiş olduğumuz başarı metriğidir. Aynı şekilde diğer metrikler için de aynısını yaparsak:

```python
# precision
cv_results['test_precision'].mean()
```

```python
# recall
cv_results['test_recall'].mean()
```

```python
# f1
cv_results['test_f1'].mean()
```

```python
# roc_auc
cv_results['test_roc_auc'].mean()
```

Accuracy değerimiz %77 iyi bir değer, fakat elimizdeki veri setinde bağımlı değişken dengeli bir şekilde dağılmadığından dolayı sadece accuracy'e göre değerlendirmemiz yanlış olur. Dolayısıyla f1 score'una da bakmamız gerekir. Buradaki f1 değerimiz ise 0.63. Değerlendirecek olursak, modelimiz accuracy değerinin belirttiği kadar iyi tahminler oluşturmuyor fakat kötü bir model de değil. Değişken mühendisliğinin de yapılmadığını düşünürsek ortalama bir model çıkardık diyebiliriz.
